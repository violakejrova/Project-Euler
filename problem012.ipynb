```python
import math

def delitele(cislo):
  pocet = 0
  odmocnina = int(math.sqrt(cislo)) #celočíselná odmocnina

  for i in range (1,odmocnina +1):
    if cislo % i == 0:
      if cislo / i == i: #aby např. 6*6 nezapočítalo číslo 6 dvakrát
        pocet += 1
      else:
        pocet += 2
  return pocet

triagnum = 0
krok = 1
while delitele(triagnum) <= 500:
  triagnum += krok
  krok +=1
print((triagnum))
```
